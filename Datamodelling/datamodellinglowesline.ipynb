```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import statsmodels.formula.api as sm
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('video_games.xlsx')
df.head()
```

```python
x = df['Review_Score']
y = df['Sales']
```

```python
plt.scatter(x, y)
plt.xlabel('Review score')
plt.ylabel('Sales')
plt.show()
```

```python
m, b = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, m*x + b, color='red')
plt.title('Scatterplot with Line of Best Fit')
plt.xlabel('Review score')
plt.ylabel('Sales')
plt.show()
#this is cool but not what im looking for i think
```

```python
lowessplot = sns.regplot(x = x, y = y, lowess = True, line_kws={"color":"red"} ) #changing color makes the whole graph change colors. Line_kws fixex the color, rest of the graph is goated
plt.show(lowessplot)
```
