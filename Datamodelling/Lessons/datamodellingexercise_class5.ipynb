```python
dfm = pd.read_csv('insurance_missings.csv')
dfm.head()
```

```python
dfm.isnull().sum()
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
ds_missing_dummies = pd.get_dummies(dfm[['sex', 'smoker']])
dfm = pd.concat([dfm, ds_missing_dummies], axis = 1)
```

```python
dfm.isnull().sum()
```
